### Spacetime coordinate axes

Recall that in PHYS 207, you were often asked to plot displacement as a function of time. For example, for a baseball thrown directly upward you would plot $y(t)$, and for a block on a horizontal spring you would plot $x(t)$. Here we will examine how the displacement and time axes "scissor together" at relativistic speeds.

The convention in relativity is to reverse the axes: $ct$ is vertical and $x$ is horizontal. Time moves forward from the bottom to the top of the plot:

```python
plt.figure(figsize=(4,4))
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.annotate(text='', xy=(-1,0), xytext=(1,0), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='', xy=(0,-1), xytext=(0,1), arrowprops=dict(arrowstyle='<->'))
plt.text(0.98, 0.04, r"$+x$", horizontalalignment='center', verticalalignment='bottom')
plt.text(-0.98, 0.04, r"$-x$", horizontalalignment='center', verticalalignment='bottom')
plt.text(-0.04, 1.02, r"$+ct$", horizontalalignment='right', verticalalignment='center')
plt.text(-0.04, -1.02, r"$-ct$", horizontalalignment='right', verticalalignment='center')
plt.axis('off')
```

Imagine that frame $K^{\prime}$ is moving in the $+x$ direction at speed $v$ relative to frame $K$. You are at rest in frame $K$. Let's figure out how the set of $(x^{\prime}, ct^{\prime})$ axes would look to you when the frames' origins $O$ and $O^{\prime}$ coincide.

<ol>
    <li>Use the fact that $x^{\prime} = 0$ on the $ct^{\prime}$ axis to define a line $ct(x)$ in frame $K$ that follows the $ct^{\prime}$ axis. Write a function that computes points on the line. $v$ will be one of your function inputs.</li>
</ol>

```python
def ct_prime_axis(x,v):
    #Gives the ct values for the ct axis when x' = 0
    return (constants.c/v) * x
```

<ol start=2>
    <li>Use the fact that $ct^{\prime} = 0$ on the $x^{\prime}$ axis to define a line $ct(x)$ that traces the $x^{\prime}$ axis in frame $K$. Write a second function that computes points on this line.</li>
</ol>

```python
def x_prime_axis(x,v):
    #Gives the ct values for the x axis when ct' = 0
    return (v/constants.c) * x
```

Use your functions from problems 1 and 2 to overplot the $ct^{\prime}$ and $x^{\prime}$ axes <strong>as a function of $x$</strong> for $v = 0.2c$.
```python
# Set up an array of independent variables x that spans (-1e10, 1e10) m
x = np.linspace(0, 2e10, num=201, endpoint=True) - 1e10

# Make a figure
fig, ax = plt.subplots(figsize=(7,7))

# YOUR CODE HERE: Use ax.plot() to plot the ct' and x' axes below
ax.plot(x, ct_prime_axis(x,v), color = "red")
ax.plot(x,x_prime_axis(x,v), color = "red")

# Move the left and bottom "spines" (borders of the plot) to the center
# This makes a math textbook-like set of (x,ct) axes instead of a box around the plot
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

# Take out the top and right spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Make the two axes have the same scale
ax.set_aspect('equal')

# Set the axes ranges
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])

# YOUR CODE HERE: Use ax.text() to add RED labels to the ct' and x' axes below
ax.text(1.05e10, ct_prime_axis(1.05e10, v), r"$ct'$", color='red', fontsize=14, verticalalignment='bottom')
ax.text(1.05e10, x_prime_axis(1.05e10, v), r"$x'$", color='red', fontsize=14, verticalalignment='bottom')
ax.text(1.02e10, 0, r"$x$ (m)", color='black', fontsize=14, verticalalignment='bottom')
ax.text(0, 1.02e10, r"$ct$ (m)", color='black', fontsize=14, horizontalalignment='right')

plt.show()
```

    <li>Make a figure that plots $ct^{\prime}$ and $x^{\prime}$ axes for $v = 0.25c$, $v = 0.5c$, and $v = 0.75c$. Use different colors for each pair of axes. Make a legend that shows which value of $v$ corresponds to each color. Label the $(x, ct)$ axes and add units.</li>

```python
x = np.linspace(-1e10, 1e10, 201)

vfrac1 = 0.25
vfrac2 = 0.5
vfrac3 = 0.75
color1 = 'purple'
color2 = 'blue'
color3 = 'pink'

fig, ax = plt.subplots(figsize=(8,8))

v1 = vfrac1 * constants.c
v2 = vfrac2 * constants.c
v3 = vfrac3 * constants.c

ax.plot(x, ct_prime_axis(x, v1), color=color1, linestyle='-', linewidth=2, label=f"ct' axis, v = 0.25c")
ax.plot(x, x_prime_axis(x, v1), color=color1, linestyle='--', linewidth=2, label=f"x' axis, v = 0.25c")

ax.plot(x, ct_prime_axis(x, v2), color=color2, linestyle='-', linewidth=2, label=f"ct' axis, v = 0.5c")
ax.plot(x, x_prime_axis(x, v2), color=color2, linestyle='--', linewidth=2, label=f"x' axis, v = 0.5c")

ax.plot(x, ct_prime_axis(x, v3), color=color3, linestyle='-', linewidth=2, label=f"ct' axis, v = 0.75c")
ax.plot(x, x_prime_axis(x, v3), color=color3, linestyle='--', linewidth=2, label=f"x' axis, v = 0.75c")

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.set_aspect('equal')
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])

ax.text(1.02e10, 0, r"$x$ (m)", color='black', fontsize=14, verticalalignment='bottom')
ax.text(0, 1.02e10, r"$ct$ (m)", color='black', fontsize=14, horizontalalignment='right')

ax.legend(loc=2, fontsize=10)

plt.show()
```

Create a similar plot to problem 4, but this time plot $(x^{\prime}, ct^{\prime})$ axes for $v = 0.5, 0.75, 0.875, 0.9375$.
```python
#On the x'-axis, when ct' = 0, the lines tilt upwards as v increases in value
#On the ct'-axis, when x' = 0, the lines tilt to the right as v increases in value
x = np.linspace(-1e10, 1e10, 201)

vfrac1 = 0.5
vfrac2 = 0.75
vfrac3 = 0.875
vfrac4 = 0.9375

color1 = "0.7"
color2 = "0.5"
color3 = "0.3"
color4 = "0.1"

fig, ax = plt.subplots(figsize=(8,8))

v1 = vfrac1 * constants.c
v2 = vfrac2 * constants.c
v3 = vfrac3 * constants.c
v4 = vfrac4 * constants.c

ax.plot(x, ct_prime_axis(x, v1), color=color1, linestyle='-', linewidth=2, label=f"ct' axis, v = 0.5c")
ax.plot(x, x_prime_axis(x, v1), color=color1, linestyle='--', linewidth=2, label=f"x' axis, v = 0.5c")

ax.plot(x, ct_prime_axis(x, v2), color=color2, linestyle='-', linewidth=2, label=f"ct' axis, v = 0.75c")
ax.plot(x, x_prime_axis(x, v2), color=color2, linestyle='--', linewidth=2, label=f"x' axis, v = 0.75c")

ax.plot(x, ct_prime_axis(x, v3), color=color3, linestyle='-', linewidth=2, label=f"ct' axis, v = 0.875c")
ax.plot(x, x_prime_axis(x, v3), color=color3, linestyle='--', linewidth=2, label=f"x' axis, v = 0.875c")

ax.plot(x, ct_prime_axis(x, v4), color=color4, linestyle='-', linewidth=2, label=f"ct' axis, v = 0.9375c")
ax.plot(x, x_prime_axis(x, v4), color=color4, linestyle='--', linewidth=2, label=f"x' axis, v = 0.9375c")

ax.plot(x, x, color='black', linestyle=':', linewidth=2, label='Light cone (limit as v → c)')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.set_aspect('equal')
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])

ax.text(1.02e10, 0, r"$x$ (m)", color='black', fontsize=14, verticalalignment='bottom')
ax.text(0, 1.02e10, r"$ct$ (m)", color='black', fontsize=14, horizontalalignment='right')

ax.legend(fontsize=10, loc=2)

plt.show()
```

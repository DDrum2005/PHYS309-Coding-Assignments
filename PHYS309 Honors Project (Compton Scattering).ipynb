# PHYS309 Tutorial: Compton Scattering

In this tutorial, I will explore Compton scattering, which is a key physics experiment that demonstrated how light carries momentum and behaves like a particle. I will derive the Compton shift formula and visualize how photon wavelength and energy change after scattering.

## 1. Background Info

Compton scattering is the description of the collision between a photon (X-ray or gamma ray) and a (basically) free electron.

Before the collision:
- The photon has energy: $E = \frac{hc}{\lambda}$
- The electron is treated as initially at rest in its own reference (rest) frame, which can be used as the invariant frame for the interaction

After the collision:
- The photon scatters at an angle, $\theta$, with an increased wavelength
- The electron recoils and carries kinetic energy with it

This process demonstrates conservation of:
- Energy
- Momentum (in 2 dimensions)

## 2. Assumptions

To derive the Compton formula, we have to assume:
- The electron is treated as initially at rest in its own reference (rest) frame
- The electron is free (its binding energy is negligible)
- Relativistic energy and momentum conservation are applied in the derivation; however, the scattered electron is assumed to remain non-relativistic after the collision
- Photon momentum is $p = \frac{h}{\lambda}$

## 3. Compton Wavelength Shift

Applying conservation of energy and momentum gives:

$\lambda' - \lambda = \frac{h}{m_e c}(1 - \cos\theta)\$

where:
- $\lambda$ = initial wavelength
- $\lambda'$ = scattered wavelength
- $\theta$ = scattering electron angle
- $\frac{h}{m_e c} = 2.43 \times 10^{-12} \, \text{m}$ (Compton wavelength of electron)

## 3.1. Special Note
- $\phi$ = scattered electron angle
- $\theta$ is measured between the incident and scattered photon directions, whereas $\phi$ is measured between the incident photon direction and the scattered electron


Source: https://www.facebook.com/100086691939822/posts/compton-scattering-compton-scattering-is-a-fundamental-phenomenon-in-modern-phys/910923901807333/

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

#constants
h = constants.h
m_e = constants.e
c = constants.c

lambda_Compton = h/(m_e * c) #Compton wavelength

theta = np.linspace(0, np.pi, 300)

delta_lambda = lambda_Compton * (1 - np.cos(theta)) #Wavelength shift

plt.figure(figsize=(8,6))
plt.plot(np.degrees(theta), delta_lambda * 1e12)

plt.title("Compton Wavelength Shift vs Scattering Angle")
plt.xlabel("Photon Scattering Angle θ (degrees)")
plt.ylabel("Wavelength Shift Δλ (pm)")
plt.show()
```

## What Does This Plot Mean?

- At $\theta = 0^\circ$, there is no scattering → $\Delta \lambda = 0$
- At $\theta = 180^\circ$, that is the maximum wavelength shift
- This plot shows that the wavelength shift depends only on the angle, not on the initial wavelength

## 4. Photon Energy After Scattering

We can also express the result in terms of photon energy:

$E' = \frac{E}{1 + \frac{E}{m_e c^2}(1 - \cos\theta)}
\$

This equation shows that when colliding, higher-energy photons lose more energy to the electron.

```python
E_0 = 100e3 * 1.602e-19 #100 keV photon in joules

mc_squared = m_e * c**2

theta = np.linspace(0, np.pi, 300)

E_prime = E_0 / (1 + (E_0/mc_squared)*(1 - np.cos(theta))) #Photon Energy

plt.figure(figsize=(8,6))
plt.plot(np.degrees(theta), E_prime / 1.602e-19, color='purple')

plt.title("Scattered Photon Energy vs Angle")
plt.xlabel("Photon Scattering Angle θ (degrees)")
plt.ylabel("Photon Energy E' (keV)")
plt.show()
```

## What Does This Plot Mean?

- Photons lose energy when scattering off electrons
- Larger angles mean that there is a larger energy transfer to the electron
- This is explicit evidence of photon momentum

```python
KE = E_0 - E_prime

plt.figure(figsize=(8,6))
plt.plot(np.degrees(theta), KE / 1.602e-19,color='hotpink')

plt.title("Electron Kinetic Energy vs Scattering Angle")
plt.xlabel("Scattering Photon Angle θ (degrees)")
plt.ylabel("Electron KE (keV)")
plt.show()
```

## 5. Conclusion
Compton scattering shows that light behaves as a particle with momentum. The wavelength shift depends only on scattering angle and provides strong evidence for quantum mechanics and relativistic energy-momentum conservation. Compton scattering also demonstrates that both photons and electrons behave like particles during the interaction, since the collision can be described using particle-like conservation of energy and momentum.

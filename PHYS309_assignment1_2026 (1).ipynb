### Questions


Recall that the speed of light in vacuum is 
\begin{equation}
c = \frac{1}{\sqrt{\epsilon_0 \mu_0}}, \; \; \; \; \; (1)
\end{equation}
where $\epsilon_0 = 8.85 \times 10^{-12}$ F/m (Farads per meter)$^{*}$ is the electric permittivity of free space and $\mu_0 = 1.26 \times 10^{-6}$ N/A$^2$ (Newtons per Ampere$^2$)$^{**}$ is the magnetic permeability of free space. The electric field created by a point charge $q$ is
\begin{equation}
E = \frac{1}{4 \pi \epsilon_0} \frac{q}{r^2}. \; \; \; \; (2)
\end{equation}
Suppose a charged particle is moving at some fraction of the speed of light in vacuum ($v_q = f c; \; 0 < f < 1$) as seen in reference frame $S$. Reference frame $S^{\prime}$ is co-moving with the point charge, so that $\frac{dx^{\prime}}{dt^{\prime}} = v^{\prime}_q = 0$ in frame $S^{\prime}$. To make the math easier, put $q$ at the origin of $S^{\prime}$.

To understand Einstein's motivation for developing the special theory of relativity, make the <strong>incorrect, erroneous, and wrong</strong> assumption that the Galilean velocity transformation holds true for all values of $f$:
\begin{equation}
\frac{dx^{\prime}}{dt^{\prime}} = \frac{dx}{dt} - v_q \; \; \; (3)
\end{equation}

<ol>
    <li>Assume the charged particle is an electron. Using Equation 2, calculate and plot the electric field as a function of distance $r^{\prime}$ from $q$ in frame $S^{\prime}$. Choose an x-axis range that makes sense to you and label both axes with correct units.</li>
</ol>

```python
epsilon_0 = constants.epsilon_0
pi = constants.pi
q = -constants.e
r_nm = np.arange(0.01,1.0,0.002)
r = r_nm*10**-9
E = (1/(4*pi*epsilon_0))*(q/r**2)
plt.plot(r,E, c='red')
plt.xlabel("Distance r' from q (m)")
plt.ylabel("Electric Field (N/C)")
plt.title ("Electric Field VS. Distance from q in Frame S'")
```

    <li>The Galilean velocity transformation suggests that objects can "catch up" with light, so that light in a vacuum appears to be going slower than the value of $c$ from Equation (1). Make a plot of $c^{\prime}$, the <strong>incorrect</strong> speed of light in frame $S^{\prime}$ according to the Galilean transformation, as a function of $f$. Put units on the axes where appropriate.</li>

```python
mu_0 = constants.mu_0
c = constants.c
f = np.arange(0,1.01,0.01)
v_q = f * c
gv = c - v_q
plt.plot(f,gv, c='red')
plt.xlabel("Fraction of the Speed of Light")
plt.ylabel("Speed of Light in Frame S' (m/s)")
plt.title ("Speed of Light in Frame S' VS. Fraction of the Speed of Light'")
```

<ol start=3>
    <li>In frame $S^{\prime}$, something has to give: either Equation (1) is no longer true, or the values of $\epsilon_0$ and/or $\mu_0$ have to change. Suppose Equation (1) is still true and $\mu_0 = \mu_0^{\prime}$. Make a plot with two vertically stacked subplots. The top subplot should show $\epsilon_0^{\prime}$ as a function of $f$, while the bottom plot should show the electric field $E^{\prime} = \frac{1}{4 \pi \epsilon_0^{\prime}} \frac{q}{{r^{\prime}}^2}$ at $r^{\prime} = 1 \mu \mathrm{m} = 10^{-6}$ m as a function of $f$.
```python
c = constants.c
mu_0 = constants.mu_0
pi = constants.pi
q = -constants.e
f = np.arange(0.1,1,0.01)
v_q = f * c
gv = c - v_q
c_prime = gv
epsilon_prime = 1/((c_prime**2)*mu_0)
r_prime = 1*10**-6
E_prime = (1/(4*pi*epsilon_prime))*(q/r_prime**2)
fig, (ax1,ax2) = plt.subplots(2,1,figsize=(8,10))
ax1.plot(f,epsilon_prime, c='red')
ax1.set_xlabel("Fraction of the Speed of Light")
ax1.set_ylabel("Electric Permittivity of Free Space (F/m)", size=12)
ax1.set_title("Electric Permittivity of Free Space in Frame S' VS. Fraction of the Speed of Light")
ax2.plot(f,E_prime, c ="blue")
ax2.set_xlabel("Fraction of the Speed of Light")
ax2.set_ylabel("Electric Field (N/C)")
ax2.set_title("Electric Field in Frame S' VS. Fraction of the Speed of Light")
plt.subplots_adjust(hspace=0.5)
```

```python
c = constants.c
mu_0 = constants.mu_0
pi = constants.pi
q = -constants.e
epsilon_0 = constants.epsilon_0
f = 0.5
v_q = f * c
gv = c - v_q
c_prime = gv
epsilon_prime = 1/((c_prime**2)*mu_0)
r_nm = np.arange(0.01,1.0,0.002)
r_prime2 = r_nm*10**-9
E_0= (1/(4*pi*epsilon_0))*(q/r_prime2**2)
E_prime = (1/(4*pi*epsilon_prime))*(q/r_prime2**2)
plt.plot(r_prime2,E_0, c='red', label = "Problem 1")
plt.plot(r_prime2,E_prime, c ="blue", label = "Problem 3")
plt.legend(loc = 4)
plt.xlabel("Distance r' (meter)")
plt.ylabel("Electric Field E' (N/C)")
plt.title("Electric Field in Frame S' VS. Distance r' when f = 0.5")
#For f = 0.5, the electric field calculated from using the epsilon prime value for both problems differ. 
#Both fields have the same shape but have slightly different electric field magnitudes, even though they are supposed to describe and 
#visualize the same situation happening. 
#These contradicting results shows that the Galilean velocity transformation cannot be used with electromagnetism
```
